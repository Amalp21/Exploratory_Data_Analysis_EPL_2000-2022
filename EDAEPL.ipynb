1. What team committed the most fouls?

-That is, for each team, which club committed the most fouls?   
-This is answered by finding out the average fouls made by each club in the years 2000â€“2022 for the home and away teams

```python
#subsetting required columns

fouls_data = football[['DateTime', 'HomeTeam', 'AwayTeam', 'HF', 'AF' ]]
fouls_data.loc[:, 'DateTime'] = pd.to_datetime(fouls_data['DateTime']).dt.year
#print(fouls_data)

#To find the clubs who have commited the most fouls in the home team
home_fouls = fouls_data[['HomeTeam', 'HF']]
home_fouls = home_fouls.groupby('HomeTeam')['HF'].agg(avg_fouls_home = 'mean').round(2).sort_values(by = 'avg_fouls_home', ascending = False)

#To find the clubs who have commited the most fouls in the away team
away_fouls = fouls_data[['AwayTeam', 'AF']]
away_fouls = away_fouls.groupby('AwayTeam')['AF'].agg(avg_fouls_away = 'mean').round(2).sort_values(by = 'avg_fouls_away', ascending = False)

#Selects the teams who made the most fouls in the home and away team
print(home_fouls.head(5))
print(away_fouls.head(5))
```

The results above show the top 5 teams in the home and away teams with the most fouls.
Key Insights:   
-Leeds is consistently foul-prone, with a high average both at home (14.01) and away (13.81).   
-Coventry stands out for having the highest average fouls when playing away (14.84), but a lower average at home (13.00).   
-There is a general trend of high fouls for these teams, with differences between their home and away performances, but all teams listed in both tables tend to commit more than 13 fouls per game on average.

3. For a team winning at half time, how does it change at full time?


```python
# Subset relevant columns
football_df = football[['FTHG', 'FTAG', 'HTHG', 'HTAG']]

# To verify what teams were winning at half time
football_df.loc[:, 'half_time_home'] = (football_df['HTHG'] > football_df['HTAG']).astype(int)
football_df.loc[:, 'half_time_away'] = (football_df['HTHG'] < football_df['HTAG']).astype(int)

# To verify what teams won at full time
football_df.loc[:, 'full_time_home'] = (football_df['FTHG'] > football_df['FTAG']).astype(int)
football_df.loc[:, 'full_time_away'] = (football_df['FTHG'] < football_df['FTAG']).astype(int)

# Determine what home teams that won at half time won or didn't win by full time
football_df['outcome_at_full_time_home'] = 0  # Default for 'Did not win'
football_df.loc[football_df['half_time_home'] == football_df['full_time_home'], 'outcome_at_full_time_home'] = 1  # 'Won'

# Determine what away teams that won at half time won or didn't win by full time
football_df['outcome_at_full_time_away'] = 0  # Default for 'Did not win'
football_df.loc[football_df['half_time_away'] == football_df['full_time_away'], 'outcome_at_full_time_away'] = 1  # 'Won'

# Visualize the outcomes for home teams
outcome_counts_home = football_df['outcome_at_full_time_home'].value_counts()
outcome_counts_home.plot(kind='bar', color=['Red', 'Green'])
plt.title('Outcome at Full Time for Home Teams Winning at Half Time')
plt.xlabel('Outcome at Full Time')
plt.ylabel('Count')
plt.xticks([0, 1], ['Did Not Win', 'Won'], rotation=0)
plt.grid(True, linewidth=0.3)
plt.show()
```

Context: Between 2000 and 2022, there were 8288 games played.   
-The visual above shows that for home teams winning at halftime, more than 6000 games at full time did not end in a win.

```python
outcome_counts_away = football_df['outcome_at_full_time_away'].value_counts()
outcome_counts_away.plot(kind='bar', color=['coral', 'blue'])
plt.title('Outcome at Full Time for Away Teams Winning at Half Time')
plt.xlabel('Outcome at Full Time')
plt.ylabel('Count')
plt.xticks([0, 1], ['Did not Win', 'Won'], rotation=0)
plt.grid(True, linewidth = -0.1)
plt.show()
```

For the away teams it was discovered that 6500+ games winning at half time did not end up in a win.

4. Does the number of red cards a team receives have an effect on its probability of winning a game?

```python
#Map the outcome to numerical values (1 for home win, 0 for draw, -1 for away win)
football['Outcome'] = football['FTR'].map({'H': 1, 'D':0, 'A': -1})

#Create a dataframe with relevant columns
correlation_data = football[['Outcome', 'HR', 'AR']]

#Compute correlation matrix
correlation_matrix = correlation_data.corr()
print(correlation_matrix)

#Create a heatmap of the correlation matrix
plt.figure(figsize = (10,6))
sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm', linewidth = 0.5)
plt.title('Correlation plot: Red Cards vs. Match outcome')
plt.show()
```

Outcome represents the final results, whether it is a home win, draw, or away win.  
HR = number of red cards by the home team.
AR = number of red cards by the away team

The heatmap above shows correlation coefficients (r) between variables. Our interest lies in the 'Outcome and AR' & 'Outcome and HR' correlation coefficients.

Outcome and AR correlation coefficient = 0.096   
-A coefficient close to zero suggests a very weak and almost negligible linear relationship between the number of red cards received by the away team and the match outcome.        
-This also indicates that, based on linear correlation alone, there is little evidence to suggest a strong association between the number of red cards received by the away team and the probability of winning the game.

Outcome and HR correlation coefficient = - 0.13     
-This indicates a negative linear relationship between the number of red cards received by the home team and the outcome of the game.   
-This weak negative association means that there is a slight tendency for more red cards from the home team to be associated with a lower probability of winning, but again, the probability is not strong.

5. Win/Loss Ratios by Team

```python
import matplotlib.pyplot as plt

# Calculate win/loss ratios
home_wins = football[football['FTR'] == 'H']['HomeTeam'].value_counts()
away_wins = football[football['FTR'] == 'A']['AwayTeam'].value_counts()
draws = football[football['FTR'] == 'D']['HomeTeam'].value_counts()

# Combine into a DataFrame
win_loss_ratio = pd.DataFrame({
    'Home Wins': home_wins,
    'Away Wins': away_wins,
    'Draws': draws
}).fillna(0)

# Calculate ratios
win_loss_ratio['Total Matches'] = win_loss_ratio['Home Wins'] + win_loss_ratio['Away Wins'] + win_loss_ratio['Draws']
win_loss_ratio['Win Ratio'] = win_loss_ratio['Home Wins'] / win_loss_ratio['Total Matches']

# Visualization
plt.figure(figsize=(12, 6))  # Adjust the figure size
win_loss_ratio['Win Ratio'].plot(kind='bar', width=0.6, color='indigo')  # Set width for bars
plt.title('Win Ratio by Team')
plt.ylabel('Win Ratio')
plt.xticks(rotation=45, ha='right')  # Rotate team names for better visibility
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

Here are the win ratios summarized :

-Top 5 Highest Win Ratios: Arsenal, Manchester United, Liverpool, Chelsea, Manchester City.  
-Bottom 5 Win Ratios: Derby, Blackpool, Coventry, Huddersfield, Cardiff.     
-5 Teams Close to the Middle: Burnley, Everton, Fulham, Leeds, Newcast.  

6. Scoring Trends

```python
# Total goals per season
goals_per_season = football.groupby('Season')[['FTHG', 'FTAG']].sum().reset_index()
goals_per_season['Total Goals'] = goals_per_season['FTHG'] + goals_per_season['FTAG']

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(goals_per_season['Season'], goals_per_season['Total Goals'], marker='o', color='b')

# Add labels for each point
for i in range(len(goals_per_season)):
    plt.text(goals_per_season['Season'][i], goals_per_season['Total Goals'][i], 
             goals_per_season['Total Goals'][i], ha='center', va='bottom')

# Title and labels
plt.title('Total Goals Per Season')
plt.xlabel('Season')
plt.ylabel('Total Goals')

# Show the plot with exact numbers
plt.grid(True)
plt.xticks(rotation=90)  # Rotating x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()

```

7. Aggregate Rivalry Stats

```python
# Group matches by team pairings (home vs. away) and aggregate rivalry statistics
rivalry_df = football.groupby(['HomeTeam', 'AwayTeam']).agg(
    total_matches=('FTHG', 'size'),
    total_home_goals=('FTHG', 'sum'),
    total_away_goals=('FTAG', 'sum'),
    total_yellow_cards_home=('HY', 'sum'),
    total_yellow_cards_away=('AY', 'sum'),
    total_red_cards_home=('HR', 'sum'),
    total_red_cards_away=('AR', 'sum'),
    total_draws=('FTR', lambda x: (x == 'D').sum())
).reset_index()

# Sum the total goals, yellow cards, and red cards
rivalry_df['total_goals'] = rivalry_df['total_home_goals'] + rivalry_df['total_away_goals']
rivalry_df['total_yellow_cards'] = rivalry_df['total_yellow_cards_home'] + rivalry_df['total_yellow_cards_away']
rivalry_df['total_red_cards'] = rivalry_df['total_red_cards_home'] + rivalry_df['total_red_cards_away']

# Combine Home and Away team into a single rivalry identifier
rivalry_df['Rivalry'] = rivalry_df['HomeTeam'] + ' vs ' + rivalry_df['AwayTeam']

# Add additional metrics
rivalry_df['goals_per_match'] = rivalry_df['total_goals'] / rivalry_df['total_matches']
rivalry_df['cards_per_match'] = (rivalry_df['total_yellow_cards'] + rivalry_df['total_red_cards']) / rivalry_df['total_matches']

# Sort rivalries by intensity (e.g., goals per match, red cards, etc.)
rivalry_df_sorted = rivalry_df.sort_values(by=['total_matches', 'goals_per_match', 'cards_per_match'], ascending=False)

# Display top rivalries
top_rivalries = rivalry_df_sorted[['Rivalry', 'total_matches', 'total_goals', 'total_yellow_cards', 'total_red_cards', 'goals_per_match', 'cards_per_match']].head(10)

print(top_rivalries)

# Plotting the most electric rivalries based on total goals and cards
top_rivalries.plot(x='Rivalry', y=['goals_per_match', 'cards_per_match'], kind='bar', figsize=(12, 6), title='Most Electric Rivalries (Goals & Cards per Match)')
plt.ylabel('Per Match Average')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
